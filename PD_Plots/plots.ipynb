```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(t, s, title, xlab='Time (h)'):
    fig, ax = plt.subplots()
    ax.plot(t, s)

    ax.set(xlabel=xlab, ylabel='Cost',
           title=title)
    ax.grid()

    #fig.savefig("test.png")
    plt.savefig(title+".eps")
```

```python
# Hores que hauria de fer - Hores que fa
t = np.arange(-50.0, 50.0, 0.01)
s = np.power(t, 2)
plot(t,s, "Saldo horari")
```

```python
# Assignatures per professor
t = np.arange(0.0, 20.0, 0.01)
s = np.power(t, 3)
plot(t,s, "Assignatures per professor", "#Assignatures")
```

```python
# Difèrencia temporal entre final i començament de classes en un mateix dia
t = np.arange(0, 10*60, 1)
s = np.concatenate([np.power(2*(15-t[:15]), 2), t[15:5*60]-t[15] , 30*np.log(t[5*60:])+t[5*60]-t[15]-30*np.log(t[5*60])])
plot(t, s, "Temps entre classes","Time (min)")
```

```python
# Hores per dia
t = np.arange(0, 10*60, 0.1)
s = np.concatenate([np.zeros(5000), np.power(t[:1000], 2)])
plot(t, s, "Hores de classe en un dia", "Time (min)");
```

```python
rooms=['Dilluns','Dimarts', 'Dimecres', 'Dijous', 'Divendres']
colors=['pink', 'lightgreen', 'lightblue', 'wheat', 'salmon']

input_files=['data1.txt']
day_labels=['Day 1']


for input_file, day_label in zip(input_files, day_labels):
    fig=plt.figure(figsize=(10,5.89))
    for line in open(input_file, 'r'):
        data=line.split()
        event=data[-1]
        data=list(map(float, data[:-1]))
        room=data[0]-0.48
        start=data[1]+data[2]/60
        end=start+data[3]/60
        # plot event
        plt.fill_between([room, room+0.96], [start, start], [end,end], color=colors[ord(event[0])-ord('A')], edgecolor='k', linewidth=0.5)
        plt.fill_between([room+0.96/2-0.01, room+0.96/2+0.01], [start, start], [end,end], color=colors[ord(event[0])-ord('A')], edgecolor='k', linewidth=0.5)
        # plot beginning time
        plt.text(room+0.02, start+0.05 ,'{0}:{1:0>2}'.format(int(data[1]),int(data[2])), va='top', fontsize=7)
        # plot event name
        plt.text(room+0.48, (start+end)*0.5, event, ha='center', va='center', fontsize=11)

    # Set Axis
    ax=fig.add_subplot(111)
    ax.yaxis.grid()
    ax.set_xlim(0.5,len(rooms)+0.5)
    ax.set_ylim(15.1, 8.9)
    ax.set_xticks(range(1,len(rooms)+1))
    ax.set_xticklabels(rooms)
    ax.set_ylabel('Time')

    # Set Second Axis
    ax2=ax.twiny().twinx()
    ax2.set_xlim(ax.get_xlim())
    ax2.set_ylim(ax.get_ylim())
    ax2.set_xticks(ax.get_xticks())
    ax2.set_xticklabels(rooms)
    ax2.set_ylabel('Time')


    #plt.title(day_label,y=1.07)
#plt.show()
plt.savefig('horari.eps')
```
